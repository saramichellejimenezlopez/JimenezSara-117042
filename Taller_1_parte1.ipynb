```python
''' 
Punto 2. Leer un número entero y escribirlo en reversa. Por ejemplo, 461975 
será 579164. No utilizar comandos de ordenamiento. 
'''

#Pediremos al usuario que ingrese un número
num = int(input('Ingrese un número entero: '))
 
#Si el número es negativo, lo convertiremos a positivo 
if num < 0:
    num *= -1
digts_al_contrario = list()

# Se obtiene  lista con los dígitos en orden contrario.
while num != 0:
    num, residuo = divmod(num, 10)
    # En residuo queda almacenado el último dígito del contenido actual de num.
    # En num queda almacenado el cociente, es decir lo que queda de dividir de
    # forma entera lo que haya actualmente en <num>
    digts_al_contrario.append(residuo)

# Formación del nuevo número
num_al_contrario = 0

for i in range(len(digts_al_contrario)):
    # Se hacen tantos ciclos como dígitos tenga el número
    num_al_contrario += (digts_al_contrario[i]
                         * (10 ** (len(digts_al_contrario) - (i+1))))

print(f'El número ingresado, escrito en reversa es: {num_al_contrario}')
```

```python
'''
Punto 5. Leer un número entero y determinar el número de cincos que están 
consecutivos en un número. Por ejemplo, en 551235555 hay 6 cincos consecutivos.
Ahora, en 234555234512 solo hay 3 cincos consecutivos. 
'''

# Se debe condicionar el programa para que solo acepte ejecutarse si y solo si
# se ingresa un número entero int()
while True:
    try:
        num = int(input('Ingrese número entero: '))
        break
    except ValueError:
        print('Debe ingresar un número entero.')

# Creamos una variable que nos servirán para determinar ciertos valores.        
cantidad_repetida = 0

while num != 0:
    if num < 0:
        num *= -1
        # Si el número es negativo lo comvertimos a un número positivo para que
        # cumpla las condiciones del ciclo.
        
    # Analizaremos el ultimo número, el antepenultimo y trasantepenúltimo 
    # número de la lista e iremos rediciendo en un cero el número para analizar
    # el ciclo y buscar los consecutivos
    unidad = num%10
    decena = (num//10)%10
    centena = (num//100)%10
    
    # Pondremos las condiciones para agregar consecutivos 
    if unidad == decena == 5:
        if unidad == decena == centena == 5:
            cantidad_repetida += 1
        else:
            cantidad_repetida += 2
    num //= 10 
    # Al final de cada ciclo dividir el número entre 10 para reducrilo en un 
    # dígito y repetir ciclo hasta que sea 0 el número
  
print ('hay', cantidad_repetida, 'cinco(s) consecutivos.')
```

```python
''' 
Punto 6. Leer un número entero y determinar si la suma de sus dígitos es un
número de Fibonacci
'''

while True:
    try: 
        num = int(input('Ingrese un número entero: '))
        break
    except ValueError:
        print('Número no válido, debe ingresar un número entero.')
    # Condicionaremos el algoritmo a que siempre cumpla con que se debe ingresar
    # un número entero para que funcione

# Definimos una lista vacía a la cual le podremos agregar como elementos, por 
# medio de operaciones y ciclos; los dígitos del número ingresado
lista = []

while num != 0:
    if num < 0: # Si es negativo el programa convertirá a positivo para 
                # ejecutarse correctamente
        num *= -1 
    
    num, residuo = divmod(num, 10)
    # En el residuo se guarda el último dígito del número ingresado por el 
    # usuario
    lista.append(residuo)
    # Queda entonces en la lista, de forma reversa, los dígitos del número 
    # ingresado como elementos
        
    # Necesitamos la suma de los dígitos del número, así que usamos sum()
    numero = sum(lista)
print('La suma de los dígitos es', numero)

# Para establecer si la suma de los dígitos del número hace parte de un número 
# de Fibonacci, usaremos Fibonacci como una función
def Fibonacci(numero):                                       
    a = 0     # Primer número lista de Fibonacci                                                      
    b = 1     # Segundo número de la lista de Fibonacci
    fib = 0   # No ha comenzado a contar Fibonacci
    while fib < numero:  # De los dígitos que lleva ninguno es la sum(num)
        fib = a+b        # Comenzará a agregar número hasta que fib < número
        a = b
        b = fib
        if fib == numero:
            return True # Cuando el número que esté de ultimo en Fibonacci 
                        # coincida con la suma de los dígitos del (num)
    return False

if Fibonacci(numero):
    print(f'{numero} es un número de la serie Fibonacci.')             
else:
    print(f'El {numero} NO es un número de la serie Fibonacci.')
```

```python
'''
Punto 9. Hacer un programa que lea las coordenadas (x1,y1,r1) y (x2,y2,r2) que 
corresponden al centro y al radio de dos círculos. Lea un punto de coordenadas 
(a,b). Determina si (a,b) está contenido: a) dentro del círculo 1; b) dentro 
del círculo 2; c) dentro de ambos círculos; d) fuera de ambos círculos. 
Recordemos que el círculo es el lugar geométrico de los puntos del plano cuya
distancia a otro punto fijo, llamado centro, es menor o igual que una cantidad  
constante, llamada radio. 
'''

while True:
    try:
        cx1 = float(input('La coordenada en x del centro del círculo 1 es: ')) 
        cy1 = float(input('La coordenada en y del centro del círculo 1 es: '))
        cr1 = float(input('Ingrese el radio del círculo 1: '))
        cx2 = float(input('La coordenada en x del centro del círculo 2 es: '))
        cy2 = float(input('La coordenada en y del centro delncírculo 2 es: '))
        cr2 = float(input('Ingrese el radio del círculo 2: '))
        a = float(input('Digite la coordenada en x de un punto: '))
        b = float(input('Digite la coordenada en y de un punto: '))
        # Se han ingresado los valores para poder definir cada círculo.
        break                                                                        
    except ValueError:                                                               
        print('Número inválido, intente de nuevo.')                   

A = ((cx1 - cr1) < a < (cx1 + cr1)) and ((cy1 - cr1) < b < (cy1 + cr1))   
    # Definimos los límites del círculo 1.

B = ((cx2 - cr2) < a < (cx2 + cr2)) and ((cy2 - cr2) < b < (cy2 + cr2))           
    # Definimos límites del círculo 2.

if A:
    print(f'El punto de coordenadas ({a}, {b}) está dentro del círculo 1.')  
if B:
    print(f'El punto de coordenadas ({a}, {b}) está dentro del círculo 2.')  

if A and B:
    print(f'El punto de coordenadas ({a}, {b}), están en ambos círculos.')
    
else:
    print(f'El punto de coordenadas ({a}, {b}), no está en ningun círculo.')
```

```python
'''
Punto 14. Leer una cadena de texto y organice alfabéticamente cada una de las 
letras que la componen, repitiendo cada una tantas veces como se encuentra. 
Por ejemplo, la cadena ‘tarea importante’ será ‘aaaeeimnoprrttt’. (Note que no 
se incluyen los espacios). 
'''

# Debemos condicionar al usuario a que solo ingrese una cadena de texto.
while True:
    try:
        texto = input('Ingrese una cadena de texto: ')
        if type(texto) == str:
            break
    except ValueError:
        print('No es una cadena de texto válido.')
        
orden_alfab = ''   # Creamos una cadena vacía en la cual ordenaremos la cadena 
                   # de texto ingresada de forma alfabetica
ordenado = sorted(texto)  # El comando que nos va a ayudar alfabeticamente es
                          # soted () 
                   
while ' ' in  ordenado:
    ordenado.remove(' ')
    # En el caso que tengamos algún espacio en la cadena de texto ingresado, 
    # eliminaremos el espacio porque así nos lo pide el ejercicio.
    
# Agregaremos a la cadena de orden_alfab cada letra que pasó por el comando 
# sorted() y así iremos formando la nueva cadena de texto
for i in range (len(ordenado)):
    orden_alfab += ordenado[i]
    
print('La cadena de texto ordenada albaféticamente es:', orden_alfab)
```
